```python
from bs4 import BeautifulSoup
```

```python
guitarists_link = []
n = 0
with open('./data/guiatrEquipment.txt') as f:
    for i in f:
#         url = f.readline().strip()
        resp = requests.get(i, headers=headers)
        print(resp)
        html = BeautifulSoup(resp.text)
        row = html.select('li[class="span2"]')
        n += len(row)
        print('Guitarist: ',n)
        for single in row:
            link = 'http://equipboard.com{}'.format(single.select_one('a')['href'])
            guitarists_link.append(link)
```
